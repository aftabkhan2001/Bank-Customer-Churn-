# BANK CUSTOMER CHURN PREDICTION 

**Bank customer churn refers to the phenomenon where customers discontinue their relationship with a bank or financial institution. This can occur for various reasons, such as dissatisfaction with services, better offers from competitors, changes in financial circumstances, or life events. Customer churn is a critical metric for banks as it directly impacts revenue and profitability. To mitigate churn, banks employ strategies such as improving customer service, offering personalized incentives, analyzing customer feedback, and leveraging data analytics to identify at-risk customers and proactively address their needs.**

# IMPORTANCE

***Revenue Impact:*** Churn directly affects a bank's revenue by reducing income from deposits, loans, and other services.

***Cost Consideration:*** Acquiring new customers is costlier than retaining existing ones, making churn management essential for cost efficiency.

***ustomer Lifetime Value (CLTV):*** Retaining customers increases their lifetime value to the bank, contributing more to long-term profitability.

***Market Reputation:*** High churn rates can indicate underlying issues and impact the bank's reputation negatively.

***Data Insights:*** Analyzing churn provides valuable insights into customer behavior, enabling banks to improve products and services.

***Competitive Edge:*** Effective churn management helps banks differentiate themselves and maintain a loyal customer base, enhancing their competitive position.

### OVER SAMPLING (SMOTE)
**--------------------------------------------------**

**SMOTE (Synthetic Minority Over-sampling Technique) is a resampling technique used in machine learning to address class imbalance by generating synthetic examples for the minority class. By creating synthetic instances along the line segments connecting existing minority class samples, SMOTE enhances the training dataset, improving the model's ability to generalize and make more accurate predictions for the minority class.**

```python
HTML('<img src="1_EonSiza89JUI4W_uZggalg.png" style="width:1200px;">')
```

```python
# CREATE A OBJECT OF SMOTE
os=SMOTE(random_state=70)

# APPLYING SMOTE TECHNIQUES 
X_train_os,y_train_os=os.fit_resample(X_train,y_train)
```

```python
# DISTRIBUTION OF MY TARGET VARIABLE("CHURN") AFTER APPLYING SMOTE TECHNIQUES 
y_train_os.value_counts()
```

***Purpose:*** Over-sampling in machine learning is a fix for imbalanced classes, where one class is way more common than the other, preventing fair predictions.

***Issue:*** Class imbalance can cause bias, making the model favor the majority class.

***Solution:*** Over-sampling tackles this by boosting the number of instances in the underrepresented class.

***Objective:*** The goal is to help the model better understand patterns in the less common class.

***Result:*** This approach enhances the model's ability to predict accurately for both the dominant and minority classes.

# ROC Curves & AUC For Each Classifier After Hyperparameter Tuning

The **ROC-AUC score is like a report card for a model in a binary test (yes or no)**. It checks how well the model can separate positive from negative things. **The ROC curve is a graph that shows how good the model is at making this separation. The higher the ROC-AUC score, the better the model is doing overall.**

```python
# COMPUTE ROC CURVES AND AUC FOR EACH CLASSIFIER
fpr_log_hyper, tpr_log_hyper, _ =metrics.roc_curve(y_test,y_pred_proba_log)
fpr_nb_hyper, tpr_nb_hyper, _=metrics.roc_curve(y_test,y_pred_proba_nb)
fpr_sv_hyper, tpr_sv_hyper, _=metrics.roc_curve(y_test,y_pred_proba_sv)
fpr_rf_hyper, tpr_rf_hyper, _ =metrics.roc_curve(y_test,y_pred_proba_rf)
fpr_xgb_hyper, tpr_xgb_hyper, _ =metrics.roc_curve(y_test,y_pred_proba_xgb)

auc_log_hyper =metrics.auc(fpr_log_hyper, tpr_log_hyper)
auc_nb_hyper =metrics.auc(fpr_nb_hyper, tpr_nb_hyper)
auc_sv_hyper =metrics.auc(fpr_sv_hyper, tpr_sv_hyper)
auc_rf_hyper =metrics.auc(fpr_rf_hyper, tpr_rf_hyper)
auc_xgb_hyper =metrics.auc(fpr_xgb_hyper, tpr_xgb_hyper)
```

```python
# PLOT ROC CURVES
plt.figure(figsize=(8, 8), dpi=100)

plt.plot(fpr_log_hyper, tpr_log_hyper, marker='.', label=f"Logistic Regression (AUC = {auc_log_hyper:.3f})")
plt.plot(fpr_nb_hyper, tpr_nb_hyper, marker='.', label=f"Naive Bayes (AUC = {auc_nb_hyper:.3f})")
plt.plot(fpr_sv_hyper, tpr_sv_hyper,marker='.',label=f"Support Vector Machine (AUC = {auc_sv_hyper:.3f})")
plt.plot(fpr_rf_hyper, tpr_rf_hyper, marker='.', label=f"Random Forest (AUC = {auc_rf_hyper:.3f})")
plt.plot(fpr_xgb_hyper, tpr_xgb_hyper, marker='.', label=f"XGBoost (AUC = {auc_xgb_hyper:.3f})")

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.title("ROC Curves For Different Classifiers")
plt.show()
```
